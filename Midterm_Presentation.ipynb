# Using new code to compare "Strawberries' price on the farms and in New York state in 1999 and 2019

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = data.head(10)['farmprice']
y = data.head(10)['newyorkretail']
```

```python
plt.scatter(x, y)
plt.xlabel('farmprice')
plt.ylabel('newyorkretail')
plt.title('Comparing Strawberries price on farm and in New York state in 2019')
```

```python
plt.scatter(x,y)
plt.xlabel('farmprice')
plt.ylabel('newyorkretail')
plt.title('Comparing Strawberries price on farm and in New York state in 1999')
```
